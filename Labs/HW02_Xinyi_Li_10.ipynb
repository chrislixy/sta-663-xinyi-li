**Exercise 4 [20 pts]**. Normalize the given matrix $M$ so that all rows sum to 1.0 (as in Exercise 2). This can then be considered as a transition matrix $P$ for a Markov chain. Find the stationary distribution of this matrix in the following ways using `numpy` and `numpy.linalg` (or `scipy.linalg`):

- By repeated matrix multiplication of a random probabilty vector $v$ (a row vector normalized to sum to 1.0) with $P$ using matrix multiplication with `np.dot`.
- By raising the matrix $P$ to some large power unitl it doesn't change with higher powers (see `np.linalg.matrix_power`) and then calculating $vP$
- From the equation for stationarity $wP = w$, we can see that $w$ must be a left eigenvector of $P$ with eigenvalue $1$ (Note: np.linalg.eig returns the right eigenvectors, but the left eighenvector of a matrix is the right eigenvector of the transposed matrix). Use this to find $w$ using `np.linalg.eig`.
- Suppose $w = (w_1, w_2, w_3)$. Then from $wP = w$, we have:
\begin{align}
w_1 P_{11} + w_2 P_{21} + w_3 P_{31} &= w_1 \\
w_1 P_{12} + w_2 P_{22} + w_3 P_{32} &= w_2 \\
w_1 P_{13} + w_2 P_{23} + w_3 P_{331} &= w_3 \\
\end{align}
This is a singular system, but we also know that $w_1 + w_2 + w_3 = 1$. Use these facts to set up a linear system of equations that can be solved with `np.linalg.solve` to find $w$.

Given matrix $M$
```python
      [[7, 8, 8],
       [1, 3, 8],
       [9, 2, 1]]
```

````python
# Your code here

# create matrix M
M=np.array([7, 8, 8, 1, 3, 8, 9, 2, 1]).reshape(3,3).astype(float)
print M, "\n"

# normalize M by row sum to get transition matrix P.
rsum=M.sum(axis=1)
print rsum
P=M / rsum[:, np.newaxis]
print P
print P.sum(axis=1), "\n"

# create a random vector v
v=np.random.ranf(3)
v=v/sum(v)
print v, "\n"



# By repeated multiplication
# make a copy of v
u=v
for i in range(50):
    u=np.dot(u,P)
print u
u_test=np.dot(u,P)
print u_test
print "Since u and u_test are the same, ", u, "is the stationary distribution of this matrix. (Method: repeated multiplication) \n"

# By raising the power of P.
P50=np.linalg.matrix_power(P,50)
P51=np.linalg.matrix_power(P,51)
print P50
print P51
print "The stationary distribution of P is ",np.dot(v,P50), ". (Method: raising P to a large power ) \n"

# By formula wP=w
left_eig=np.linalg.eig(P.T)[1][:,0]
print "Left Eigenvector is", left_eig
w=np.real(left_eig/sum(left_eig))
print "The stationary distribution of P is ",w, ". (Method: using formula wP=w) \n"

# By solving linear equations
# a is coefficient matrix of the linear system.
# x=(w1,w2,1-w1-w2). By subtracting a by the corresponding coefficient, can get a determined b (b not consist of unknowns)
# for example, w1*P11+w2*P21+w3*P31=w1 can be writen as w1*(P11-1)+w2*P21+w3*P31=0
a=np.array([[0.3043-1, 0.0833, 0.75],[0.3478, 0.25-1, 0.1667],[0.3478+1, 0.6667+1, 0.0833]])
print a
b=np.array([0,0,1])
print b
x=np.linalg.solve(a,b)
print "The stationary distribution of P is ",x, ". (Method: solving linear system of equations)"
````
